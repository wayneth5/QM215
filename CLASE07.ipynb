```python
def generaDias(nEstudiantes):
  dias= np.random.randint(1,366, size=nEstudiantes)
  op=False
  for i in range(0,nEstudiantes):
    n1=dias[i]
    v=dias[i+1:]
    for j in range(0,len(v)):
      if v[j]==n1:
        op=True
  return op
```

```python
#Calcular una probabilidad
n=0
for i in range(1000):
  if generaDias(20)==True:
    n+=1
print(n/1000*100)
```

Reacción Química A--> P

image.png

```python
A=1.0
P=0.0
k=0.1
t=0
dt=0.1
tt=[]
AA=[]
PP=[]
while t<50:
  tt.append(t)
  AA.append(A)
  PP.append(P)
  dA=-k*A*dt
  dP=k*A*dt
  P=P+dP
  A=A+dA
  t+=dt
tt=np.array(tt)
AA=np.array(AA)
PP=np.array(PP)

#Decaimiento exponencial

x=np.arange(0,50,5) #rango de valores entre 0 y 50
y=np.exp(-k*x) #funcion exponencial
y2=1-y
plt.scatter(x,y,color="red",s=20)
plt.scatter(x,y2,color="blue",s=20)
plt.plot(tt,AA, color="fuchsia")
plt.plot(tt,PP, color="skyblue")
plt.xlabel("t(seg)")
plt.ylabel("[A]")
plt.grid(True)
plt.title("Reacción de primer orden A-->P")
plt.show()
```

#Reacción de segundo orden
A+A --> P

```python
A=1.0
P=0.0
k=0.1
t=0
dt=0.1
tt=[]
AA=[]
PP=[]
rojo_puro = '#FF0000'
azul_puro = '#0000FF'

while t<50:
  tt.append(t)
  AA.append(A)
  PP.append(P)
  dA=-k*A*A*dt
  dP=k*A*A*dt
  P=P+dP
  A=A+dA
  t+=dt
tt=np.array(tt)
AA=np.array(AA)
PP=np.array(PP)

x=np.arange(0,50,5)
Ao=1.0
invAt=k*x+1/Ao
y=1/invAt
y2=1-y
plt.scatter(x,y,color="black",s=20)
plt.scatter(x,y2,color="black",s=20)

plt.plot(tt,AA, color=rojo_puro)
plt.plot(tt,PP, color=azul_puro)
plt.xlabel("t(seg)")
plt.ylabel("[A]")
plt.grid(True)
plt.title("Reacción de Segundo orden A-->P")
plt.show()
```

```python
A=1.0
P=0.0
X=0.0
k1=0.1
k2=0.2
t=0
dt=0.1
tt=[]
AA=[]
PP=[]
XX=[]
rojo_puro = '#FF0000'
azul_puro = '#0000FF'

while t<50:
  tt.append(t)
  AA.append(A)
  PP.append(P)
  XX.append(X)
  dA=-k1*A*dt
  dX=k1*A*dt-k2*X*dt
  dP=k2*X*dt
  P=P+dP
  A=A+dA
  t+=dt
  X=X+dX

tt=np.array(tt)
AA=np.array(AA)
PP=np.array(PP)
XX=np.array(XX)

plt.plot(tt,AA)
plt.plot(tt,XX)
plt.plot(tt,PP)
plt.xlabel("t(seg)")
plt.ylabel("[A]")
plt.grid(True)
plt.title("Reacción consecutiva orden A-->X    x--> p ")
plt.show()
```

Movimiento Browniano


```python
for i in range(0,10):
    dx=np.random.normal(0,0.1,size=1000)
    dy=np.random.normal(0,0.1,size=1000)
    x=np.cumsum(dx)
    y=np.cumsum(dy)
    plt.plot(x,y)
plt.show()
```
